```python
df2=df.drop(columns=['Date','Day'])   #removing date and day from the dataframe.
df2
```

```python
t=pd.date_range(start='1/01/2021',periods=1000)    #getting dates for inserting in a table.
t
```

```python
df2.set_index(t)       #setting date as an index.
```

```python
from datetime import datetime
def h():                             #defining function to get weekdays from the given dates
    weekday=list()
    for i in t:
        r=i.weekday()
        weekday.append(r)
    return weekday

    
h()
```

```python
df2.insert(loc=1,column="weekday",value=h(),allow_duplicates=False)   #inserting weekday with new dates.
df2
```
